
# üìä Salary Dataset Analysis (Exam-Oriented)

**Dataset:** Annual salaries (in $1000) of 20 employees in a tech company.

This notebook explains each statistical measure clearly, suitable for **exam answers**.



## 1Ô∏è‚É£ Dataset

We are given the following sorted salary data (in $1000):

45, 48, 50, 52, 55, 58, 60, 62, 65, 68,  
70, 72, 75, 78, 80, 85, 90, 95, 150, 500


```

import numpy as np

data = np.array([45, 48, 50, 52, 55, 58, 60, 62, 65, 68,
                 70, 72, 75, 78, 80, 85, 90, 95, 150, 500])
data
```


## 2Ô∏è‚É£ Range

**Formula:**  
Range = Maximum value ‚àí Minimum value

It measures the spread of the data.


```

data_range = data.max() - data.min()
data_range
```


## 3Ô∏è‚É£ Mean, Variance & Standard Deviation

### Mean
The mean is the average of all observations.

### Variance (Sample)
Variance measures how far values are spread from the mean.  
We use **sample variance** because this is a sample of employees.

**Formula:**  
\[ s^2 = \frac{\sum (x_i - \bar{x})^2}{n-1} \]

### Standard Deviation
Standard deviation is the square root of variance.


```

mean = data.mean()
variance = data.var(ddof=1)   # sample variance
std_dev = data.std(ddof=1)    # sample standard deviation

mean, variance, std_dev
```


## 4Ô∏è‚É£ Quartiles

- **Q1**: First quartile (25% data below)
- **Q2**: Median (50% data below)
- **Q3**: Third quartile (75% data below)

Since the data is already sorted and n = 20 (even),
the median is the average of the 10th and 11th values.


```

Q1 = np.percentile(data, 25, method='midpoint')
Q2 = np.percentile(data, 50, method='midpoint')
Q3 = np.percentile(data, 75, method='midpoint')

Q1, Q2, Q3
```


## 5Ô∏è‚É£ Interquartile Range (IQR)

**Formula:**  
IQR = Q3 ‚àí Q1

It measures the spread of the middle 50% of the data.


```

IQR = Q3 - Q1
IQR
```


## 6Ô∏è‚É£ Outlier Detection Using IQR Method

**Lower Fence:**  
\[ Q1 - 1.5 \times IQR \]

**Upper Fence:**  
\[ Q3 + 1.5 \times IQR \]

Any value outside these fences is considered an outlier.


```

lower_fence = Q1 - 1.5 * IQR
upper_fence = Q3 + 1.5 * IQR

lower_fence, upper_fence
```


## 7Ô∏è‚É£ Identifying Outliers

Values smaller than the lower fence or larger than the upper fence are outliers.


```

outliers = data[(data < lower_fence) | (data > upper_fence)]
outliers, len(outliers)
```
